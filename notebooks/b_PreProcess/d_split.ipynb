<center><h1>Split</h1></center>

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
features = df.drop(columns=['net_price', 'qtym'])
target = df.net_price
```

```python
splitter = TimeSeriesSplit(n_splits=3)
folds = list(splitter.split(features, target))
```

```python
folds[0]
```

```python
folds[1]
```

```python
d_folds = {}
fold = 0
for train, test in folds:
    d_folds[f'train_{fold}'] = train
    d_folds[f'test_{fold}'] = test
    fold += 1
```

```python
d_folds
```

```python
np.savez('d:/demand-forecast-SQGroup/data/folds.npz', **d_folds)
```
