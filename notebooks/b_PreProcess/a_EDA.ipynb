# Univariate EDA

## Continuous

### summary stats

```python
df.describe().T
```

### distributions

```python
df.select_dtypes(['float64']).columns
```

```python
# Continuous Feature distribution
cols = list(df.select_dtypes(['float64']).columns)[2:] +['population',]
fig, grid = plt.subplots(3, 2, figsize=(18, 20))
colId = 0
colorPallette = ['set1', 'set2', 'set3']
for row in range(3):
    for col in range(2):
        feature = cols[colId]
        sns.histplot(data=df, ax=grid[row, col], x=feature, label=feature, bins=30)
        grid[row, col].set_title(f'{feature} DIstribution')
        colId += 1
plt.tight_layout()
plt.show()
```

```python
# Targets distribution
fig, grid = plt.subplots(1, 2, figsize=(18, 10))
sns.histplot(data=df, ax=grid[0], x='Net_Price', label='Net_Price', bins=30)
grid[0].set_title(f'Net_Price DIstribution')
sns.histplot(data=df, ax=grid[1], x='qtym', label='qtym', bins=30)
grid[1].set_title(f'qtym DIstribution')
plt.tight_layout()
plt.show()
```

```python
df.select_dtypes('int64').columns
```

```python
# Discrete Feature distribution
cols = ['item_type_id', 'wire', 'noc', 'tlcolt']
fig, grid = plt.subplots(2, 2, figsize=(18, 20))
colId = 0

for row in range(2):
    for col in range(2):
        feature = cols[colId]
        sns.countplot(data=df, ax=grid[row, col], x=feature, label=feature)
        grid[row, col].set_title(f'{feature} Distribution')
        colId += 1        
plt.tight_layout()
plt.show()
```

```python
#
cols =['population', 'outlet_id', 'division']
fig, grid = plt.subplots(1, 3, figsize=(16, 16))

for col in range(3):
        feature = cols[col]
        sns.countplot(data=df, ax=grid[col], y=feature, label=feature)
        grid[col].set_title(f'{feature} Distribution')     
plt.tight_layout()
plt.show()
```

### Univariate - Categorical

```python
df.select_dtypes('object').columns
```

```python

# Categorical Feature distribution
cols =  ['item_type', 'size', 'Category', 'grade', 'ecoInd', 'pcx', 'uses', 'appliation_group']
fig, grid = plt.subplots(4, 2, figsize=(18, 28))
colId = 0

for row in range(4):
    for col in range(2):
        feature = cols[colId]
        sns.countplot(data=df, ax=grid[row, col], y=feature, label=feature)
        grid[row, col].set_title(f'{feature} Distribution')
        colId += 1        
plt.tight_layout()
plt.show()
```
