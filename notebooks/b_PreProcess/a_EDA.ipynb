```python
# Continuous Feature distribution
cols = ['rm', 'dfc', 'Area_(km)^2', 'pcx', 'exach', 'tmtm','tms_cr', 'mas', 'kpi', 'Literacy_Rate_(%)','Population_(Approx.)',]
fig, grid = plt.subplots(6, 2, figsize=(18, 20))
colId = 0
colorPallette = ['set1', 'set2', 'set3']
for row in range(6):
    for col in range(2):
        feature = cols[colId]
        sns.histplot(data=df, ax=grid[row, col], x=feature, label=feature, bins=30)
        grid[row, col].set_title(f'{feature} DIstribution')
        colId += 1
        if colId == 11:
            break
plt.tight_layout()
plt.show()
```

```python
# Targets distribution
fig, grid = plt.subplots(1, 2, figsize=(18, 10))
sns.histplot(data=df, ax=grid[0], x='Net_Price', label='Net_Price', bins=3)
grid[0].set_title(f'Net_Price DIstribution')
sns.histplot(data=df, ax=grid[1], x='qtym', label='qtym', bins=3)
grid[1].set_title(f'qtym DIstribution')
plt.tight_layout()
plt.show()
```

```python

# Categorical Feature distribution
cols = ['Category', 'Base_Size', 'client_id', 'Grade', 'Uses',
       'Application_Group', 'ecoInd', 'division', 'mkt']
fig, grid = plt.subplots(5, 2, figsize=(18, 28))
colId = 0

for row in range(4):
    for col in range(2):
        feature = cols[colId]
        sns.countplot(data=df, ax=grid[row, col], y=feature, label=feature)
        grid[row, col].set_title(f'{feature} Distribution')
        colId += 1 
        if colId == 9:
            break
plt.tight_layout()
plt.show()
```
